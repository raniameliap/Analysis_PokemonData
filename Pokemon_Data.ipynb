**DATA CLEANING**

```python
df = df.drop(['#'], axis=1)
df
```

```python
#Check for missing values
missing_values = df.isnull().sum()
print("Missing values before cleaning:\n", missing_values)
```

```python
#Verify that there are no more missing values
missing_values_after_cleaning = df.isnull().sum()
print("Missing values after cleaning:\n", missing_values_after_cleaning)
```

```python
#Check for duplicates and remove if necessary
duplicate_rows = df.duplicated().sum()
print("Number of duplicate rows:", duplicate_rows)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
```

```python
# One-hot encoding untuk 'Type 1' dan 'Type 2'
df_encoded = pd.get_dummies(df, columns=['Type 1', 'Type 2'], drop_first=True)

# Normalisasi kolom numerik
numeric_columns = df_encoded.columns[3:-2]
scaler = StandardScaler()
df_encoded[numeric_columns] = scaler.fit_transform(df_encoded[numeric_columns])
```

```python
# Distribusi Tipe Pokemon
plt.figure(figsize=(12, 6))
sns.countplot(x='Type 1', data=df, order=df['Type 1'].value_counts().index)
plt.title('Distribution of Pokemon Primary Types')
plt.xticks(rotation=45)
plt.show()
```

```python
#Hubungan antara HP dan Attack
plt.figure(figsize=(12, 6))
sns.scatterplot(x='HP', y='Attack', data=df, hue='Type 1')
plt.title('Relationship between HP and Attack')
plt.show()
```

```python
#Histogram Total Statistik
plt.figure(figsize=(12, 6))
sns.histplot(df['Total'], bins=20, kde=True)
plt.title('Histogram of Total Stats')
plt.show()
```

```python
#Boxplot Total Statistik berdasarkan Generasi
plt.figure(figsize=(12, 6))
sns.boxplot(x='Generation', y='Total', data=df)
plt.title('Boxplot of Total Stats by Generation')
plt.show()
```
