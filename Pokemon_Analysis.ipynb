**ANALYSIS & VISUALIZATION**

```python
# Calculate the number of each generation
Generation = df['Generation'].value_counts()
Generation
```

```python
# Calculate the number of each Legendary
Legendary = df['Legendary'].value_counts()
Legendary
```

```python
#  Total Pokemon for each Type 1
type1_total = df.groupby('Type 1')['Total'].count()
type1_total
```

```python
# Total Pokemon for each Type 2
type2_total = df.groupby('Type 2')['Total'].count()
type2_total
```

```python
#Apakah ada perbedaan signifikan dalam jumlah Pokémon legendaris dan non-legendaris yang memiliki nilai untuk Attack dan Defense?
Pokemon_Legendary= df.groupby('Legendary',)[['Attack','Defense']].count()
Pokemon_Legendary
```

```python
#Top 3 Strongest Pokemon dengan Menggunakan Total Pada Type 1
Generation_pokemon1 = df.nlargest(3, 'Total', keep='first')['Type 1'] # Apply nlargest to the entire DataFrame and select 'Type 1' column
Generation_pokemon1
```

```python
#Top 3 Strongest Pokemon based on Total value in Generation 1
Generation_pokemon1 = df[df['Generation'] == 1].nlargest(3, 'Total')
Generation_pokemon1
```

```python
# Plot
plt.figure(figsize=(10, 6))
type1_total.sort_values().plot(kind='bar', color='orange')
plt.title('Total Pokemon for each Type 1')
plt.xlabel('Type 1')
plt.ylabel('Total Statistik')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Plot
plt.figure(figsize=(10, 6))
type2_total.sort_values().plot(kind='bar', color='yellow')
plt.title('Total Pokemon for each Type 2')
plt.xlabel('Type 2')
plt.ylabel('Total Statistik')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# pie chart
plt.figure(figsize=(10, 8))
plt.pie(Generation, labels=Generation.index, autopct='%1.1f%%', startangle=120, colors=['pink','yellow','orange','green','skyblue','red'])
plt.title('Percentage of Pokemon Based on Generation')
plt.axis('equal')
plt.show()
```

```python
# pie chart
plt.figure(figsize=(9, 8))
plt.pie(Legendary, labels=Legendary.index, autopct='%1.1f%%', startangle=140, colors=['yellow','orange'])
plt.title('Percentage of Pokemon Based on Legendary')
plt.axis('equal')
plt.show()
```

```python
corr = df[['Total', 'HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed', 'Generation']].corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
```

```python
# Membuat bar plot
plt.figure(figsize=(12, 8))
sns.barplot(x='Total', y='Name', data=strongest_pokemon, palette='viridis')
plt.title('Top 5 Pokemon with the Strongest Total Stats')
plt.xlabel('Total Stats')
plt.ylabel('Name Pokemon')
plt.show()
```

```python
# bar plot
plt.figure(figsize=(12, 8))
sns.barplot(x='Total', y='Name', data=weakest_pokemon, palette='viridis')
plt.title('Top 5 Pokemon with the Weakest Total Stats')
plt.xlabel('Total Stats')
plt.ylabel('Nama Pokemon')
plt.show()
```
